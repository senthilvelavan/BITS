```python
def appendcsv(globFiles):
    import os
    import glob
    import pandas as pd
    import numpy as np

    allFiles = glob.glob(os.path.join(globFiles))

    np_array_list = []
    for file_ in allFiles:
        df = pd.read_csv(file_,index_col=None, header=0)
        np_array_list.append(df.as_matrix())

    comb_np_array = np.vstack(np_array_list)
    big_frame = pd.DataFrame(comb_np_array)
    return big_frame
```

```python
def runAccuracy(X_train,X_test,Y_train,Y_test):
    for model in models:
        print(model)
        model.fit(X_train,Y_train)
        print(model.score(X_test,Y_test))
        Y_pred = model.predict(X_test)
        print(le_winner.inverse_transform(Y_pred))
        print(classification_report(Y_test,Y_pred))
        print(confusion_matrix(Y_test,Y_pred))
        accuracy.append(accuracy_score(Y_test,Y_pred))
    print(pd.DataFrame({'modelNames':modelNames,'models':models,'accuracy':accuracy}))
```

```python
def labelEncodeAndXY(venues_df):
    from sklearn.preprocessing import LabelEncoder
    le_date = LabelEncoder()
    le_winner = LabelEncoder()
    le_team1 = LabelEncoder()
    le_team2 = LabelEncoder()
    le_winType = LabelEncoder()
    le_mom = LabelEncoder()
    le_tossWinner = LabelEncoder()
    le_tossDecision = LabelEncoder()
    le_city = LabelEncoder()
    venues_df['date'] = le_date.fit_transform(venues_df['date'])
    venues_df['winner'] = le_winner.fit_transform(venues_df['winner'])
    venues_df['team1'] = le_team1.fit_transform(venues_df['team1'])
    venues_df['team2'] = le_team2.fit_transform(venues_df['team2'])
    venues_df['winType'] = le_winType.fit_transform(venues_df['winType'])
    venues_df['ManOfMatch'] = le_mom.fit_transform(venues_df['ManOfMatch'])
    venues_df['tossWinner'] = le_tossWinner.fit_transform(venues_df['tossWinner'])
    venues_df['tossDecision'] = le_tossDecision.fit_transform(venues_df['tossDecision'])
    venues_df['city'] = le_city.fit_transform(venues_df['city'])
    venues_df['winMargin'] = venues_df['winMargin'].astype(int)

    X = venues_df[['date','team1','team2','winType','winMargin','ManOfMatch','tossWinner','tossDecision','city']]
    Y = venues_df['winner']

    from sklearn.model_selection import train_test_split
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
    
    return X_train,X_test,Y_train,Y_test
```

```python
X_train,X_test,Y_train,Y_test = labelEncodeAndXY(venuesHyd_df)
runAccuracy(X_train,X_test,Y_train,Y_test)
```

```python
def labelEncodeAndXY_Mom(venues_df):
    print(venues_df['winner'].value_counts())
    from sklearn.preprocessing import LabelEncoder
    le_date1 = LabelEncoder()
    le_winner1 = LabelEncoder()
    le_team11 = LabelEncoder()
    le_team21 = LabelEncoder()
    le_winType1 = LabelEncoder()
    le_mom1 = LabelEncoder()
    le_tossWinner1 = LabelEncoder()
    le_tossDecision1 = LabelEncoder()
    le_city1 = LabelEncoder()
    venues_df['date'] = le_date1.fit_transform(venues_df['date'])
    venues_df['winner'] = le_winner1.fit_transform(venues_df['winner'])
    print(le_winner1.classes_)
    venues_df['team1'] = le_team11.fit_transform(venues_df['team1'])
    venues_df['team2'] = le_team21.fit_transform(venues_df['team2'])
    venues_df['winType'] = le_winType1.fit_transform(venues_df['winType'])
    #venues_df['ManOfMatch'] = le_mom.fit_transform(venues_df['ManOfMatch'])
    venues_df['tossWinner'] = le_tossWinner1.fit_transform(venues_df['tossWinner'])
    venues_df['tossDecision'] = le_tossDecision1.fit_transform(venues_df['tossDecision'])
    #venues_df['city'] = le_city.fit_transform(venues_df['city'])
    venues_df['winMargin'] = venues_df['winMargin'].astype(int)

    X = venues_df[['date','team1','team2','winType','winMargin','tossWinner','tossDecision']]
    Y = venues_df['winner']

    from sklearn.model_selection import train_test_split
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
    
    return X_train,X_test,Y_train,Y_test, le_winner1
```
